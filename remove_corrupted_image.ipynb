```python
img_dir="/home/rahul/imagenet_downloader/ImageNet-downloader/person/"

corrupt_img_dir="/home/rahul/imagenet_downloader/ImageNet-downloader/person_corrupted/"
#good_img_dir="/home/rahul/imagenet_downloader/ImageNet-downloader/forklift_good/"
```

```python
from PIL import Image
import os,time
```

```python
def verify_image(img_file):
     #test image
     try:
        v_image = Image.open(img_file)
        v_image.verify()
        return True;
        #is valid
        #print("valid file: "+img_file)
     except OSError:
        return False;
```

```python
#main script
for root, dirs, files in os.walk(img_dir):
    for file in files:
        if file.lower().endswith(".jpg"):
             currentFile=os.path.join(root, file)
             #test image
             if verify_image(currentFile):
                 #new_file_name=good_img_dir+time.strftime("%Y%m%d%H%M%S_"+os.path.basename(currentFile))
                 print("good file,keeping ")
                 
             else:
                 new_file_name=corrupt_img_dir + os.path.basename(currentFile)
                 print("corrupted, moving to dir: "+new_file_name)
                 try:
                     os.rename(currentFile, new_file_name)
                 except WindowsError:
                     print("error moving file")
```
