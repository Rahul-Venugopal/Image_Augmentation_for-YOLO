OPEN ALL RENAMED IMAGES AND RMOVE CORRUPTED FILES

```python
def verify_image(img_file):
     #test image
     try:
        v_image = Image.open(img_file)
        v_image.verify()
        return True;
        #is valid
        #print("valid file: "+img_file)
     except OSError:
        return False;
```

```python
corrupt_img_dir = os.path.dirname(os.path.dirname(img_dir))+'/'+'corrupt_images/'
for root, dirs, files in os.walk(img_dir):
    for file in files:
        
             currentFile=os.path.join(root, file)
             #test image
             if verify_image(currentFile):
                 #new_file_name=good_img_dir+time.strftime("%Y%m%d%H%M%S_"+os.path.basename(currentFile))
                 print("good file,keeping ")
                 
             else:
                 new_file_name=corrupt_img_dir + os.path.basename(currentFile)
                 print("corrupted, moving to dir: "+new_file_name)
                 if not os.path.exists(corrupt_img_dir):
                      os.makedirs(corrupt_img_dir)
                 try:
                     os.rename(currentFile, new_file_name)
                 except WindowsError:
                     print("error moving file")
```

```python
image_ids = open(img_list_file).read().strip().split()
print("Number of images in the image list in %s is %s " %(img_list_file , len(image_ids)))
n_without_bb = 0   # flag for counting images without bb
n_with_bb = 0      # # flag for counting images with bb
```

```python
print("Matching annotations in %s with images in %s" %(annotation_dir,img_dir))
for image_id in image_ids:
    
  annotation_id = image_id.replace(".jpg",".xml")
  print (image_id)
  annotation_file = os.path.join('%s%s' % (annotation_dir,annotation_id))
  print("searching for %s" %(annotation_file))
  exists = os.path.isfile(annotation_file)
  currentFile=os.path.join('%s%s' % (img_dir,image_id))
  #new_file_name=img_with_bb_dir + image_id
  if exists:
      #print("Annotation exists")
      new_file_name=img_with_bb_dir + image_id
      print("Annotation_Exists , moving to dir: " + new_file_name)
      if not os.path.exists(img_with_bb_dir):
          os.makedirs(img_with_bb_dir)
      try:
          os.rename(currentFile, new_file_name)
          print("moved")
          n_with_bb = n_with_bb + 1
          #print("here is the problem")
      except:
            print("error moving an image with bounding box")
  else:
      #print("No annotation file named %s found at %s" %(annotation_id , annotation_dir))
      new_file_name = img_without_bb_dir + image_id
      #currentFile=os.path.join('%s%s' % (img_dir,image_id))
      #print(currentFile)
      #print(os.path.basename(image_id))
      #new_file_name=img_without_bb_dir + image_id
      #print(new_file_name)
      print("No bounding box available , moving to dir: " + new_file_name)
      if not os.path.exists(img_without_bb_dir):
          os.makedirs(img_without_bb_dir)
      try:
          os.rename(currentFile, new_file_name)   #to move file 
          #shutil.copyfile(currentFile, new_file_name) #to copy file
          print("COPIED")
          n_without_bb = n_without_bb + 1
      except :
             print("error moving an image without boundingbox ")
#print("Remaining images with bounding boxe = %d" %(len(image_ids)-n_without_bb))    
```
